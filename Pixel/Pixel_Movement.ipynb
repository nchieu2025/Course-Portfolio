```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Initialize figure and axis
fig, ax = plt.subplots()
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_xticks([])
ax.set_yticks([])

# Initialize a single pixel (scatter point)
pixel, = ax.plot([], [], 'ro', markersize=5)

def init():
    pixel.set_data([], [])
    return pixel,

def update(frame):
    x = frame % 10  # Move pixel horizontally
    y = (frame // 10) % 10  # Move pixel vertically if needed

    # Change: Convert x and y to lists
    pixel.set_data([x], [y])
    return pixel,

# Create animation
ani = animation.FuncAnimation(fig, update, frames=100, init_func=init, blit=True, interval=100)

# Display animation in Google Colab
from IPython.display import HTML
HTML(ani.to_jshtml())
```
